```python
import pandas as pd
df=pd.read_csv('titanic-passengers.csv', sep=";" )
df
```

```python
def preprocess_data(new_data):
    new_data['Age'].fillna(new_data['Age'].mean(),inplace=True)
    new_data.replace({'Sex':{'male': 1,'female':0}},inplace=True)
    new_data['Cabin']=new_data.Cabin.fillna('G6')
    new_data=new_data.drop(["Cabin"],axis=1)
    new_data.replace({'Survived':{'Yes': 1,'No':0}},inplace=True)
    return new_data
data=preprocess_data(df)
print(data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import tree   
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
x=data.drop(["Survived", "Name", "Ticket", "Embarked"], axis=1)
y= data["Survived"]

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20,random_state=10)

dtree = tree.DecisionTreeClassifier()  
dtree.fit(x_train, y_train)  
y_pred=dtree.predict(x_test) 
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
from sklearn.tree import plot_tree
def plot_decision_tree(tr):
    global x,y
    plt.figure(figsize= (30,20))
    plot_tree(tr, filled= True,feature_names = x.columns, proportion = True)
```

```python
plot_decision_tree(dtree)
```

```python
dtree = tree.DecisionTreeClassifier(max_depth=4)  
dtree.fit(x_train, y_train)  
y_pred=dtree.predict(x_test) 
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
plot_decision_tree(dtree)
```

```python
dtree = tree.DecisionTreeClassifier(max_depth= 6,criterion="gini", splitter="random",max_leaf_nodes= 10,min_samples_leaf= 6)  
dtree.fit(x_train, y_train)  
y_pred=dtree.predict(x_test) 
print("score:{}".format(accuracy_score(y_test, y_pred)))
```

```python
clf=RandomForestClassifier(n_estimators=10) 
clf.fit(x_train, y_train) 
y_pred=clf.predict(x_test) 
print("Accuracy:", metrics.accuracy_score(y_test, y_pred)) 
```
